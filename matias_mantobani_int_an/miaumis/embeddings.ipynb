```python
from dotenv import load_dotenv
from langchain_community.embeddings import OpenAIEmbeddings
from langchain_chroma import Chroma
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter #Chunkeo

#Cargamos credenciales de entorno Api key
load_dotenv()
print()

#Cargar PDF
PATH_PDF = "./docs/gatos.pdf"
embeddings_function = OpenAIEmbeddings()
loader= PyPDFLoader(PATH_PDF)
documents = loader.load()

# Criterios de chunkeo de este documento

# Definir chunkeo con un tamaño de 1000 caracteres y un solapamiento de 100
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
# Dividir el documento en los chunks del tamano especificados
chunked_docs = text_splitter.split_documents(documents)

# Cargar documentos chunked en Chroma
Chroma.from_documents(
    documents=chunked_docs,  # Usamos los documentos chunked
    embedding=embeddings_function,
    persist_directory="docs/Chroma"
)
```

```python
# Mostrar el contenido del documento original
print("Contenido del documento original:")
for doc in documents:
    print(doc.page_content)  # Esto muestra el contenido de cada página del PDF
```

```python
# Imprimir el número total de chunks creados a partir de los documentos
print(f"Number of chunks: {len(chunked_docs)}")

# Iterar sobre cada chunk creado y enumerarlo para imprimir su contenido
for i, chunk in enumerate(chunked_docs):
    # Imprimir el índice del chunk y su contenido
    print(f'Chunk {i}: {chunk}')
    print(f"Chunk {i} length: {len(chunk.page_content)}")
```
